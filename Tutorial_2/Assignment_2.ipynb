```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

For a given dataset: (1,1.2), (2,1.9), (3,3.2). Find the line which fits the data using maximum likelihood function. Plot the line with the given dataset. Please take beta = 1; therefore you need to find only w (i.e. intercept and slope only).

```python
dataset = [[1,1.2], [2, 1.9], [3, 3.2]]
beta = 1

n = len(dataset)
x = np.array([lst[0] for lst in dataset])
y = np.array([lst[1] for lst in dataset])
```

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("DATASET")
plt.savefig("Datapoints.png")
plt.show()
```

```python
sigma = 1/math.sqrt(beta)

xi_square = np.array([i*i for i in x])
summation_xi = np.sum(x)
summation_yi = np.sum(y)
summation_xi_square = np.sum(xi_square)

```

```python
def get_line_parameters():
    slope = ((summation_yi*summation_xi) - (n*x.dot(y.T)))/((summation_xi*summation_xi) - (n*summation_xi_square))
    intercept = (summation_yi/n) - slope*(summation_xi/n)
    return slope, intercept
```

```python
w0, w1 = get_line_parameters()
```

```python
y_line = w0*x + w1
y_line
```

```python
plt.scatter(x, y, label = "Datapoints", color="green")
plt.plot(x, y_line, label="Best_fit_line", color = "orange")
plt.legend(["Best_fit_line","Datapoints"])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Best Fit Line")
plt.savefig("Best_line.png")
plt.show()
```
